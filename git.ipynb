#git使い方
###gitリポジトリの初期化
プロジェクトをgitで管理  
→プロジェクトのリポジトリに移動  
→＜git init＞コマンドを実行  
⇒gitリポジトリが初期化され、gitによるバージョン管理ができる  
###ファイルを管理対象に追加
→ファイルを作成  
→＜git add＞コマンドを実行  
→＜git commit＞コマンドを実行  
⇒作成したファイルをgitの管理対象に加えられる  
（例）＄ echo Hello > README.md  
　　　＄ git add README.md  
　　　＄ git commit -m "first commit"   
　　　　 ※ -mの後の文字列はそのコミットについてのメモ
###編集履歴の記録・確認
####記録
→管理対象ファイルを変更  
→＜git add＞コマンドを実行  
→＜git commit＞コマンドを実行
⇒管理対象ファイルの変更が記録される  
※＜git status＞コマンドで差分のあるファイルを確認できる  
※＜git diff＞コマンドで変更内容を確認できる  
　　※＜git add＞済みのファイルは《--cached》オプションが必要  
####確認
→＜git log＞コマンドを実行  
⇒これまで記録したコミットログが確認できる  
※黄色字で表示されているのが、コミットハッシュ(コミットを一意に識別するためのもの)
###チェックアウト
→＜git checkout コミットハッシュ＞コマンドを実行  
⇒履歴をさかのぼって特定のバージョン時点の状態にする  
※元々の状態に戻すには＜git checkout main＞コマンドを実行  
###作業のやり直し
####間違えて＜git add＞してしまったファイルを取り消す
→（例）＜git reset HEAD ファイル名＞  
→＜git reset＞  
　　※一括で取り消す
####間違えて行ったコミットの取り消し
→（例）＜git reset HEAD＞
###ブランチとマージ
####ブランチ
時系列を分岐して記録していくためのもの
####マージ
ブランチの結合  

#github使い方
###ローカルリポジトリにコミット
【git使い方-ファイルを管理対象に追加】と同様
###リモートリポジトリにプッシュ
→ローカルリポジトリ配下で＜git remote add origin https://github.com/ユーザー名/リモートリポジトリ名.git＞ コマンドを実行し、ローカルリポジトリとリモートリポジトリを紐づける    
→＜git push origin master＞コマンドでプッシュを実行  
###ブランチの作成
→＜git branch ブランチ名＞コマンドを実行  
※ブランチが正常に作成されているかは＜git branch＞で確認できる  
※＜git checkout ブランチ名＞でブランチを移動できる  
###ブランチにプッシュ
→指定のブランチに移動後、リモートリポジトリへのプッシュと同様のコマンドを実行する
###ブランチのマージ
→一般的には、メインのmasterとマージさせる  
→＜git checkout master＞  
→＜git merge 結合させるブランチ名＞  
→＜git push origin master＞githubにプッシュする  
##ブランチを削除
→＜git branch -d 削除したいブランチ名＞  
→＜git branch＞結果の確認  
→＜git push --delete origin ブランチ名＞GitHub上のリモートリポジトリからの削除
##ブランチからプルする
共同開発者が特定のブランチで開発を行う場合
→＜git checkout ブランチ名＞特定のブランチへ移動  
→＜git pull＞
